Q: Что такое решающее дерево (Decision Tree)?

A: Это алгоритм машинного обучения, который принимает решения, задавая последовательность вопросов к данным. Каждый вопрос разбивает данные на группы, пока алгоритм не придет к итоговому ответу.

Q: Из каких элементов состоит граф решающего дерева?

A: 1. Корень (Root node) — самый первый вопрос, охватывающий все данные. 2. Внутренние узлы (Internal nodes) — промежуточные проверки условий. 3. Листья (Leaf nodes) — конечные узлы, которые выдают само предсказание.

Q: Как дерево решает, какой признак выбрать для разбиения?

A: Оно перебирает все признаки и все возможные пороги (значения) для них, считая «критерий информативности» (например, Джини для классификации или MSE для регрессии). Выбирается то разбиение, которое сильнее всего уменьшает хаос (неоднородность) в данных.

Q: В чем главная уязвимость решающих деревьев?

A: Склонность к сильному переобучению (overfitting). Если не ограничивать дерево, оно будет расти до тех пор, пока в каждом листе не останется по одному объекту, просто выучив обучающую выборку наизусть.

Q: Как бороться с переобучением одного дерева?

A: Использовать гиперпараметры для ограничения роста: задать максимальную глубину (max_depth), минимальное число объектов для разбиения (min_samples_split) или минимальное число объектов в листе (min_samples_leaf).

Q: Чем отличаются задачи классификации и регрессии для листьев дерева?

A: В классификации лист выдает класс (чаще всего класс большинства объектов, попавших в этот лист). В регрессии лист выдает число (обычно среднее значение целевой переменной всех объектов в этом листе).
