### B. Пропуски и очистка
* Покажите стратегии dropna() и fillna(). Заполняйте средним, медианой, константой — и обязательно обоснуйте выбор текстовым комментарием.
* Отдельно покажите, как считаете моду (mode) и заполняете ею пропуски в категориальных колонках.

```python
# 1. Удаление колонок (drop)
# Если в колонке отсутствует больше 80-90% данных, она вряд ли поможет модели.
# Например, 'PoolQC' (качество бассейна) или 'Alley' (тип аллеи).
columns_to_drop = ['PoolQC', 'MiscFeature', 'Alley', 'Fence']
df = df.drop(columns=columns_to_drop)
print("Удалили колонки с критическим количеством пропусков.")

# 2. Удаление строк (dropna)
# Если пропусков в колонке буквально 1-2 штуки на весь датасет (например, 'Electrical'), проще удалить эти конкретные строки, чтобы не искажать статистику.
df.dropna(subset=['Electrical'], inplace=True)
print("Удалили строки с единичными пропусками.")

# 3. Заполнение константой (fillna)
# В колонке 'GarageType' (тип гаража) пропуск (NaN) означает не потерю данных, а то, что ГАРАЖА НЕТ. Заполнять медианой или модой в этом случае — логическая ошибка.
df['GarageType'] = df['GarageType'].fillna('No_Garage')
print("Заполнили пропуски гаражей константой 'No_Garage'.")

# 4. Заполнение медианой (fillna)
# Колонка 'LotFrontage' (длина фасада) — пример.
# Так как площадь участков может сильно отличаться (есть гигантские поместья — выбросы), среднее значение сильно исказится.
# Медиана гораздо устойчивее к выбросам.
median_frontage = df['LotFrontage'].median()
df['LotFrontage'] = df['LotFrontage'].fillna(median_frontage)
print(f"Заполнили LotFrontage медианой: {median_frontage}")

# 5. Заполнение средним (fillna)
# Колонка 'MasVnrArea' (площадь каменной облицовки). Распределение здесь более-менее равномерное без экстремальных выбросов, поэтому можно использовать среднее.
mean_masvnr = df['MasVnrArea'].mean()
df['MasVnrArea'] = df['MasVnrArea'].fillna(mean_masvnr)
print(f"Заполнили MasVnrArea средним: {mean_masvnr:.2f}")

# 6. Заполнение модой (fillna) для категорий
# Колонка 'MasVnrType' (тип облицовки) — это текст/категория. Нельзя посчитать среднее у текста. Поэтому мы находим самое часто встречающееся значение (моду).
# Метод .mode() возвращает Series (колонку), поэтому мы берем первый элемент [0].
mode_masvnr_type = df['MasVnrType'].mode()[0]
df['MasVnrType'] = df['MasVnrType'].fillna(mode_masvnr_type)
print(f"Заполнили MasVnrType модой: '{mode_masvnr_type}'")


print("\n--- Оставшиеся пропуски ---")
remaining_missing = df.isnull().sum()
print(remaining_missing[remaining_missing > 0].sort_values(ascending=False).head())
```

### C. Расширенная статистика
* Для числовых колонок выведите: min, max, mean, median, mode.
* Посчитайте percentile/quantile (например, 5, 25, 50, 75, 95 перцентили).

*Самостоятельно изучите и посчитайте: дисперсию (variance), асимметрию (skewness) и эксцесс (kurtosis). Попробуйте объяснить, что они значат для ваших данных.*

```python
import numpy as np
num_cols = df.select_dtypes(include=[np.number]).columns

# 1. min, max, mean, median, mode
basic_stats = pd.DataFrame({
    'Min': df[num_cols].min(),
    'Max': df[num_cols].max(),
    'Mean': df[num_cols].mean(),
    'Median': df[num_cols].median(),
    'Mode': df[num_cols].mode().iloc[0]
})

print("--- 1. Базовая статистика (первые 10 колонок) ---")
display(basic_stats.head(10))

# 2. Перцентили / Квантили
percentiles = [0.05, 0.25, 0.50, 0.75, 0.95]
quantiles_df = df[num_cols].quantile(percentiles).T

print("\n--- 2. Перцентили (первые 10 колонок) ---")
display(quantiles_df.head(10))

# 3. Variance, Skewness, Kurtosis
advanced_stats = pd.DataFrame({
    'Variance': df[num_cols].var(),
    'Skewness': df[num_cols].skew(),
    'Kurtosis': df[num_cols].kurtosis()
})

print("\n--- 3. Продвинутая статистика (первые 10 колонок) ---")
display(advanced_stats.head(10))
```

1. Дисперсия показывает, насколько сильно значения признака отклоняются от среднего значения.
Например, в LotArea есть сильный разброс данных.
2. Асимметрия показывает, насколько распределение данных скошено влево или вправо относительно нормального распределения.
К пременным, в которых видна высокая положительная ассиметрия, нужно будет применить методы, которые делают распределение более нормальным.
3. Эксцесс показывает "остроту" пика распределения и "тяжесть" его хвостов. В библиотеке Pandas используется эксцесс Фишера, где у нормального распределения эта метрика равна $0$.
Высокий эксцесс в колонке X говорит о наличии сильных выбросов - например, в LotArea.

### D. Фичи: Энкодинг и Инжиниринг (Feature Engineering)
* Сделайте кодирование категорий: OneHotEncoder (или pd.get_dummies) — обязательно.
* Если уместно: Label Encoding или Target Encoding. Покажите датафрейм до/после.

*Самостоятельно изучите: попробуйте применить Feature Hashing (Hashing Encoder) для признаков с большим числом уникальных значений.
Генерация новых фич: создайте 2-3 новых признака. Например, перемножьте/поделите две существующие колонки, вытащите месяц из даты или сгруппируйте редкие категории в одну "Other".*

```python
import pandas as pd
import category_encoders as ce

# 1. Генерация новых фичей (Feature Engineering)
print("--- 1. Генерация новых признаков ---")

# 1.1. Общая площадь дома (TotalArea)
# Складываем жилую площадь над землей и площадь подвала
df['TotalArea'] = df['GrLivArea'] + df['TotalBsmtSF'].fillna(0)
print("Создан признак 'TotalArea' (Общая площадь).")

# 1.2. Возраст дома на момент продажи (HouseAge)
# Год продажи минус год постройки
df['HouseAge'] = df['YrSold'] - df['YearBuilt']
print("Создан признак 'HouseAge' (Возраст дома).")

# 1.3. Группировка редких категорий в 'Other'
# В колонке 'SaleCondition' есть редкие статусы. Оставим только 'Normal' и 'Partial', остальное — 'Other'
popular_conditions = ['Normal', 'Partial']
df['SaleCondition_Grouped'] = df['SaleCondition'].apply(lambda x: x if x in popular_conditions else 'Other')
print(f"Уникальные значения новой фичи SaleCondition_Grouped: {df['SaleCondition_Grouped'].unique()}")

# 2. Ordinal Encoding (Для порядковых признаков)
print("\n--- 2. Ordinal Encoding ---")
df['KitchenQual'] = df['KitchenQual'].fillna('TA')

print("До маппинга (первые 5 значений):")
display(df[['KitchenQual']].head())

quality_map = {
    'Po': 1,
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}

df['KitchenQual_Ordinal'] = df['KitchenQual'].map(quality_map)
print("После маппинга:")
display(df[['KitchenQual', 'KitchenQual_Ordinal']].head())

# 3. Feature Hashing (Hashing Encoder)
print("\n--- 3. Feature Hashing ---")

print("До Hashing (колонка Neighborhood):")
display(df[['Neighborhood']].head())

hashing_enc = ce.HashingEncoder(cols=['Neighborhood'], n_components=5)
df = hashing_enc.fit_transform(df)

print("После Hashing (появились колонки col_0 ... col_4):")
hash_cols = [col for col in df.columns if 'col_' in col]
display(df[hash_cols].head())

# 4. One-Hot Encoding (OHE)
print("\n--- 4. One-Hot Encoding (pd.get_dummies) ---")
# OHE используется для номинальных признаков (где нет порядка).
# Например, тип отопления ('GasA', 'GasW' и т.д.). Каждому значению создается своя колонка с 1 или 0.

cols_for_ohe = ['CentralAir', 'Street', 'SaleCondition_Grouped']
print(f"Размер датафрейма ДО One-Hot Encoding: {df.shape}")
df = pd.get_dummies(df, columns=cols_for_ohe, drop_first=True)
print(f"Размер датафрейма ПОСЛЕ One-Hot Encoding: {df.shape}")

ohe_cols = [col for col in df.columns if any(orig_col in col for orig_col in cols_for_ohe)]
print("\nНовые сгенерированные OHE колонки:")
display(df[ohe_cols].head())
```

### E. Визуализация
Используем ВСЕ ТРИ библиотеки (matplotlib, seaborn, plotly). Не для галочки, а со смыслом!
- Распределения: гистограммы / KDE.
- Scatter plot — ищем зависимости между числами.
- Box plot (ящик с усами) — ищем аномалии/выбросы.
- Bar/count plot для категорий.
- Heatmap корреляций (напишите короткий вывод, какие колонки сильно коррелируют).
(По желанию: pairplot, violinplot, интерактивный график Plotly с параметром hover_data).

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set_theme(style="whitegrid")

print("--- 1. Распределение (Гистограмма + KDE) ---")
plt.figure(figsize=(10, 6))
sns.histplot(df['SalePrice'], kde=True, bins=40, color='blue')
plt.title('Распределение цен на дома (SalePrice)', fontsize=14)
plt.xlabel('Цена ($)')
plt.ylabel('Количество домов')
plt.show()

print("\n--- 2. Box plot (Ящик с усами) — поиск аномалий ---")
plt.figure(figsize=(10, 6))
sns.boxplot(x='KitchenQual_Ordinal', y='SalePrice', data=df, hue='KitchenQual_Ordinal', palette='Set2', legend=False)
plt.title('Влияние качества кухни на цену (и поиск выбросов)', fontsize=14)
plt.xlabel('Качество кухни (1 - Плохо, 4 - Отлично)')
plt.ylabel('Цена ($)')
plt.show()

print("\n--- 3. Scatter plot (Диаграмма рассеяния) ---")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TotalArea', y='SalePrice', data=df, alpha=0.6, color='purple')
plt.title('Зависимость цены от общей площади дома', fontsize=14)
plt.xlabel('Общая площадь (TotalArea)')
plt.ylabel('Цена ($)')
plt.show()

print("\n--- 4. Интерактивный Scatter plot (Plotly) с hover_data ---")
fig = px.scatter(df,
                 x='TotalArea',
                 y='SalePrice',
                 color='HouseAge',
                 hover_data=['YearBuilt', 'OverallQual'],
                 title='Интерактивный график: Площадь vs Цена (цвет = Возраст дома)')
fig.show()

print("\n--- 5. Bar plot (Столбчатая диаграмма) для категорий ---")
plt.figure(figsize=(8, 5))
sns.countplot(x='SaleCondition', data=df, hue='SaleCondition', palette='viridis', legend=False)
plt.title('Количество проданных домов по статусу сделки', fontsize=14)
plt.xlabel('Статус сделки (SaleCondition)')
plt.ylabel('Количество')
plt.show()

print("\n--- 6. Heatmap (Тепловая карта корреляций) ---")
plt.figure(figsize=(12, 10))

numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
top_corr_features = corr.index[abs(corr["SalePrice"]) > 0.5]
top_corr = numeric_df[top_corr_features].corr()

sns.heatmap(top_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Тепловая карта: Топ признаков, коррелирующих с ценой', fontsize=14)
plt.show()
```
